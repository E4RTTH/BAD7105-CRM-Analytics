## Data Preparation

```python
df.drop([0], inplace=True)

df.head()
```

```python
df.isnull().sum()
```

```python
df = df.fillna("ไม่")
```

```python
df.drop(columns="Timestamp", inplace=True)
```

```python
df.isnull().sum()
```

```python
def encode_units(x):

    if "ไม่" in str(x) :
      return 0
    elif "ไม่" not in str(x) :
      return 1

hot_encoded_df = df.applymap(encode_units)
```

```python
hot_encoded_df.head()
```

```python
hot_encoded_df.drop(columns="PS5", inplace=True)
```

```python
hot_encoded_df.columns
```

## Market Basket

```python
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules

frequent_itemsets = apriori(hot_encoded_df, min_support=0.02, use_colnames=True, max_len=2)
```

```python
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1)
rules.head(10)
```

```python
rules.shape
```

```python
rules_filter = rules[ (rules['lift'] >= 1) & (rules['confidence'] >= 0.5) &  (rules['support'] >= 0.2) ]

rules_filter
```

```python
## Check recommendation item
market_basket = rules_filter.sort_values(by=['lift', 'confidence', 'support', 'consequent support', 'antecedent support'], ascending=False)

market_basket.head(50)
```

```python
rules_filter.max()
```

## Collaborative Filtering

```python
item_item_matrix = pd.DataFrame(index=hot_encoded_df.columns,columns=hot_encoded_df.columns)
```

```python
from scipy.spatial.distance import cosine

for i in range(0,len(item_item_matrix.columns)) :
    # Loop through the columns for each column
    for j in range(0,len(item_item_matrix.columns)) :
      # Fill in placeholder with cosine similarities
      item_item_matrix.iloc[i,j] = 1 -cosine(hot_encoded_df.iloc[:,i],hot_encoded_df.iloc[:,j])
```

```python
links = item_item_matrix.rename_axis('related item', 
                                     axis='columns').stack().reset_index()
links.columns = ['item', 'related item','cosine sim']

links.head()
```

```python
links_filtered=links.loc[ (links['cosine sim'] > 0.1) & 
                         (links['item'] != links['related item']) ]

cf = links_filtered.sort_values(by=['cosine sim'], ascending=False)
```
